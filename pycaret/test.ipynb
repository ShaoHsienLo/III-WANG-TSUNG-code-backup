```python
from pycaret.classification import *
import pandas as pd
from lime.lime_tabular import LimeTabularExplainer
import random
```

```python
model = load_model("best_recall_model_XGBClassifier")
model
```

```python
df1 = pd.read_csv("../labeled-data/1st-2st-labeling/all-labeled-data.csv")
df2 = pd.read_csv("../labeled-data/3st-labeling/all-labeled-data.csv")
data = pd.concat([df1, df2], ignore_index=True)
data = data.drop(columns=data.filter(regex="discharge.*").columns)
data = data.drop(columns=data.filter(regex=".*_count_nonzero").columns)
data = data.drop(columns=["original label", "final label"])
data = data.reset_index(drop=True)
```

```python
len(data.columns)
```

```python
X_train = pd.read_csv("X_train.csv")
X_train
```

```python
explainer = LimeTabularExplainer(X_train.values, feature_names=X_train.columns, class_names=['A', 'B'], 
                                 mode='classification')
# i = random.randint(0, data.shape[0])
exp = explainer.explain_instance(data.iloc[0], model.predict_proba)
exp.show_in_notebook(show_table=True)
```
